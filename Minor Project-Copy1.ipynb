```python
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/NK/Desktop/MBA Sem 2 Minor Project/trip_data.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
timefmt = "%Y-%m-%d %H:%M:%S" 
data['started_at'] = pd.to_datetime(data['started_at'], format = timefmt, errors='coerce')
```

```python
data['ended_at'] = pd.to_datetime(data['ended_at'], format = timefmt, errors='coerce')
```

```python
data['ride_length'] = pd.to_datetime(data['ride_length'], format = "%H:%M:%S", errors='coerce')
```

```python
data.info()
```

```python
data.head()
```

```python
data['ride_length'] = data['ride_length'].dt.time
```

```python
data.head()
```

```python
data['day_of_week'] = data['day_of_week'].astype('object')
```

```python
data['day_of_week'].replace(1, 'Monday', inplace=True)
data['day_of_week'].replace(2, 'Tuesday', inplace=True)
data['day_of_week'].replace(3, 'Wednesday', inplace=True)
data['day_of_week'].replace(4, 'Thursday', inplace=True)
data['day_of_week'].replace(5, 'Friday', inplace=True)
data['day_of_week'].replace(6, 'Saturday', inplace=True)
data['day_of_week'].replace(7, 'Sunday', inplace=True)
```
